#**Cleaning the Data**



# 1.  ***cleaning the cdr dataset***




```python
# combining the three datasets 
mtn_data = pd.concat([df_1, df_3, df_2])
mtn_data
```

```python
# identifying duplicated columns in geo_cells ie df_5
mtn_data.duplicated()

#dropping the duplicate columns
mtn_data.drop_duplicates()
```

```python
# Checking if there are missing values in our cdr data
# this function returns a boolean value for all the cells that have Nan  value
mtn_data.isnull()
```

```python
# dropping null values
#no_missing_cdr = mtn_data.dropna()
# no_missing_cdr # commented it since it got rid of all the records.. cant work with n empty table 
 
#can't do that so instead we'll drop columns with all null values
mtn_data.dropna(axis=1, how='all')

# also dropping values with less than 8 observations
mtn_data.dropna(thresh=8)
```
